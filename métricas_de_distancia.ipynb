```python
import cv2, numpy
import cv2 as cv
import numpy as np
import math as Math
from matplotlib import pyplot as plt
from os import listdir
from os.path import isfile, join
from cmath import sqrt

def computeDistance(v1,v2):
    distance=0
    for r in range(0,3):
        distance+=(v1[r]-v2[r])**2
    distance=np.sqrt(distance)
    return(distance)

def computeDistance2(v1,v2):
    distance=0
    total = 64
    N = Math.ceil(Math.pow(total, float(1)/3))
    for r in range(0,N):
        distance+=(v1[r]-v2[r])**2
    distance=np.sqrt(distance)
    return(distance)

def chi_distance(v3, v4, eps = 1e-10):
    d = 0.5 * np.sum([((a - b) ** 2) / (a + b + eps)
        for (a, b) in zip(v3, v4)])
    return d

def Manhattan(v5,v6, eps = 1e-10):
    z = np.sum([((a - b) + eps) for (a, b) in zip(v5, v6)])
    k = abs(z)
    return k

def Batacharia(v7,v8, eps = 1e-10):
    q = abs(sqrt(np.sum([sqrt(a*b)/(sqrt(np.sum([a*b])) + eps) for (a,b) in zip(v7,v8)])))
    p = 1 - q
    x = abs(p)
    return x

def computeRGB(img):
    rows,cols,channels = img.shape
    RGB = np.zeros(3)
    for ch in range(0,3):
        acum=0
        for f in range(0,rows):
            for c in range(0,cols):
                acum+=img[f,c,ch]
        RGB[ch]= acum/(f*c)
    return(RGB)

def computehistcolor(img):
    b,g,r = cv.split(img)
    height, width, prof = img.shape
    total = 64
    N = Math.ceil(Math.pow(total, float(1)/3))
    rango = int(Math.ceil(float(256)/N))
    pixel = np.arange(total).reshape(N,N,N)
    for i in range(0,N,1):
        for j in range(0,N,1):
            for k in range(0,N,1):
                pixel[i][j][k]=0

    for i in range(0,height,1):
        for j in range(0,width,1):
            valor = pixel[int((b[i][j])/rango)] [int((g[i][j])/rango)][int(r[i][j]/rango)]
            pixel[int((b[i][j])/rango)] [int((g[i][j])/rango)][int(r[i][j]/rango)]=valor+1
            Px = pixel.flatten()
    return(Px)

#RGB Promedio

#SEARCH IMAGES
mypath='C:/Users/Ninja/Pictures/IMGPY'
onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
images = np.empty(len(onlyfiles), dtype=object)
for n in range(0, len(onlyfiles)):
    images[n] = cv.imread( join(mypath,onlyfiles[n]) )

#COMPUTE RGB
imagesRGB = np.zeros((len(images),3))
equal = np.zeros(len(images))
for i in range(0 ,len(images)):
    imagesRGB[i,:] = computeRGB(images[i])
    #equal[i]=computeDistance(computeRGB(cv.imread('C:/Users/Ninja/Pictures/IMGPY/177.jpg')),imagesRGB[i,:])
    #equal[i]=chi_distance(computeRGB(cv.imread('C:/Users/Ninja/Pictures/IMGPY/177.jpg')),imagesRGB[i,:])
    #equal[i]=Manhattan(computeRGB(cv.imread('C:/Users/Ninja/Pictures/IMGPY/177.jpg')),imagesRGB[i,:])
    #equal[i]=Batacharia(computeRGB(cv.imread('C:/Users/Ninja/Pictures/IMGPY/177.jpg')),imagesRGB[i,:])
print(imagesRGB)
print(equal)

#Histogramas
'''
total = 64
N = Math.ceil(Math.pow(total, float(1)/3))
mypath='C:/Users/Ninja/Pictures/IMGPY'
onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
images = numpy.empty(len(onlyfiles), dtype=object)
for n in range(0, len(onlyfiles)):
    images[n] = cv2.imread(join(mypath,onlyfiles[n]))

#COMPUTE COLOR HISTOGRAMS
imagesRGB = np.zeros((len(images),total))
equal = np.zeros(len(images))
for i in range(0 ,len(images)):
    imagesRGB[i,:] = computehistcolor(images[i])
    #equal[i]= computeDistance2(computehistcolor(cv2.imread('C:/Users/Ninja/Pictures/IMGPY/176.jpg')),imagesRGB[i,:])
    #equal[i]= chi_distance(computehistcolor(cv2.imread('C:/Users/Ninja/Pictures/IMGPY/176.jpg')),imagesRGB[i,:])
    #equal[i]= Manhattan(computehistcolor(cv2.imread('C:/Users/Ninja/Pictures/IMGPY/176.jpg')),imagesRGB[i,:])
    #equal[i]= Batacharia(computehistcolor(cv2.imread('C:/Users/Ninja/Pictures/IMGPY/177.jpg')),imagesRGB[i,:])
print(imagesRGB)
print(equal)
'''
'''PLOT IMAGES'''
plt.figure('RESULTADOS')
plt.subplot(31,10,3)
plt.axis("off")
plt.imshow(cv.cvtColor(cv.imread('C:/Users/Ninja/Pictures/IMGPY/177.jpg'), cv.COLOR_BGR2RGB))
upward=sorted(equal)
for s in range(6,len(images)+6):
    plt.subplot(31,10,upward.index(equal[s-6])+6)
    plt.axis("off")
    plt.imshow(cv.cvtColor(images[s-6], cv.COLOR_BGR2RGB))
plt.show()
```
